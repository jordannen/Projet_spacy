# Installation:
### 1. Vous pouvez instaler spaCy avec la commande pip:
- pip install -U pip setuptools wheel 
- pip install -U spacy
- python -m spacy download zh_core_web_sm   #moèle de mandarin
- python -m spacy download en_core_web_sm   #modèle d'anglais
- python -m spacy download fr_core_news_sm  #modèle de français
- python -m spacy download es_core_news_sm  #modèle de l'espagnol
- python -m spacy download ru_core_news_sm  #modèle du russe
- python -m spacy download hr_core_news_sm  #modèle du croate
- python -m spacy download sr_core_news_sm  #modèle du serbe
- python -m spacy download xx_ent_wiki_sm  #modèle multilangue

### 2. Vous pouvez l'installer avec la commande conda:
- conda install -c conda-forge spacy

### 3. Vous pouvez l'installer avec git si vous voulez changer le code base: 

- python -m pip install -U pip setuptools wheel # install/update build tools
- git clone https://github.com/explosion/spaCy  # clone spaCy
- cd spaCy                                      # navigate into dir
- python -m venv .env                           # create environment in .env
- source .env/bin/activate                      # activate virtual env
- pip install -r requirements.txt               # install requirements
- pip install --no-build-isolation --editable . # compile and install spaCy

### 4. Si vous voulez installer spaCy dans jupyter, il faut ajouter '!' avant la command.
- !python -m spacy download en_core_web_sm

### 5. Vous pouvez aussi valider la version du spaCy sur votre système:
- python -m spacy validate

⚠️: Veillez à redémarrer votre noyau ou votre moteur d'exécution après l'installation pour vous assurer que le paquet pipeline installé peut être trouvé.

```python
# importer des modèles :
nlp_fr = spacy.load('fr_core_news_sm') #modèle du français
#nlp_sp = spacy.load("es_core_news_sm")  #modèle de l'espagnol
nlp_en = spacy.load("en_core_web_sm") #modèle de l'anglais
#nlp_zh = spacy.load("zh_core_web_sm") #modèle du mandarin
#nlp_ru = spacy.load("ru_core_news_sm")  #modèle du russe
#nlp_hr = spacy.load("hr_core_news_sm")  #modèle du croate
#nlp_sr = spacy.load("sr_core_news_sm")  #modèle du serbe
#nlp_mlt_lg = spacy.load("xx_ent_wiki_sm") #modèle multilangue
```

```python
#textes : 
text_en = """When Sebastian Thrun started working on self-driving cars at Google in 2007, few people outside the company took him seriously. "It's worth talking about," Thrun said in an interview with Recode earlier this week."""
text_fr = """Lorsque Sebastian Thrun a commencé à travailler sur les voitures autonomes chez Google en 2007, peu de personnes en dehors de l'entreprise l'ont pris au sérieux. « Ça vaut la peine de parler », a déclaré Thrun, dans une interview avec Recode plus tôt cette semaine."""
text_zh = """2007年，当塞巴斯蒂安-瑟伦(Sebastian Thrun)开始在谷歌研究自动驾驶汽车时，公司外很少有人认真对待他。本周早些时候，Thrun在接受Recode采访时说: “这是值得讨论的问题。” """ 
text_ru = """Когда Себастьян Трун начал работать над беспилотными автомобилями в Google в 2007 году, мало кто за пределами компании воспринимал его всерьез. «Об этом стоит поговорить», — сказал Трун в интервью Recode ранее на этой неделе."""
text_hr = """Kada je Sebastian Thrun 2007. počeo raditi na samovozećim automobilima u Googleu, malo ga je ljudi izvan tvrtke shvaćalo ozbiljno. "Vrijedi o tome razgovarati", rekao je Thrun u intervjuu za Recode ranije ovog tjedna."""
text_sr_l = """Kada je Sebastijan Thrun počeo da radi na samovozećim automobilima u Guglu 2007. godine, malo ljudi van kompanije ga je shvatilo ozbiljno. „Vredi pričati o tome“, rekao je Thrun u intervjuu za Recode ranije ove nedelje."""
text_sr_c = """Када је Себастијан Трун почео да ради на самовозећим аутомобилима у Гуглу 2007. године, мало људи ван компаније га је схватило озбиљно. „Вреди причати о томе“, рекао је Трун у интервјуу за Recode раније ове недеље."""
text_esp =""" Cuando Sebastian Thrun comenzó a trabajar en automóviles autónomos en Google en 2007, pocas personas ajenas a la empresa lo tomaron en serio. "Vale la pena hablar de eso", dijo Thrun en una entrevista con Recode a principios de esta semana."""
```

# Analyse des dépendances:

```python
# Exemple en anglais:

doc = nlp_en(text_en)

print("\t\t\t racine\t\t dep\t\t tête")
for chunk in doc.noun_chunks:
    print(chunk.text, ":\t",chunk.root.text, "\t\t", chunk.root.dep_,"\t\t",chunk.root.head.text)
    print(" ")
```

```python
# Exemple en mandarin simplifie:
# The 'noun_chunks' syntax iterator is not implemented for language 'zh'.

doc = nlp_zh(text_zh)
print("\t\t\tracine  dep  tête")
for chunk in doc.noun_chunks:
    print(chunk.text, ":     ",chunk.root.text, chunk.root.dep_,chunk.root.head.text)
    print(" ")
```
